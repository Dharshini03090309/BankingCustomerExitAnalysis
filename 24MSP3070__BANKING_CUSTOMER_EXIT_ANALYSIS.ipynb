DATA CLEANING

```python
#Converting column names to lower case to avoid possible name errors
data.columns = data.columns.str.lower()
data.columns
```

```python
#Replace attrition_flag with a boolean column called "churn" as true or false
data.loc[:,'churn'] = (data['attrition_flag'] == 'Attrited Customer').astype(int)
data1 = data.drop(columns='attrition_flag')
data1.churn.value_counts()
```

```python
#Conversion to categorical columns
data1.loc[:, 'gender'] = data1.gender.astype('category')
data1.loc[:, 'education_level'] = data1.education_level.astype('category')
data1.loc[:, 'marital_status'] = data1.marital_status.astype('category')
data1.loc[:, 'income_category'] = data1.income_category.astype('category')
data1.loc[:, 'card_category'] = data1.card_category.astype('category')
```

```python
#checking for null values
data1.isna().sum()
```

```python
#checking count for each category
#count for marital_status
data1.marital_status.value_counts()
```

```python
#Imputation- filling unknown values with frequent type value method
def fill_marital_status(data1):
  data1.loc[:,'marital_status'] = data1.marital_status.replace('Unknown','Married')
  return data1

data1 = fill_marital_status(data1)
data1.marital_status.value_counts()
```

```python
#count for income_category
data1.income_category.value_counts()
```

```python
#Imputation- filling unknown with "income less that $40k"
def fill_incomes(data1):
  data1.loc[:, 'income_category'] = data1.income_category.replace('Unknown', 'Less than $40K')
  return data1

data1 = fill_incomes(data1)
data1.income_category.value_counts()
```

```python
# Pearson correlation heatmap
def plot_corr_matrix(data1):
    fig = plt.figure()
    fig.set_size_inches(11,10)
    ax = fig.add_subplot()
    numeric_data = data1.select_dtypes(include=[np.number])
    corr_mat = numeric_data.corr()
    mask = np.triu(corr_mat)
    cmap = sns.diverging_palette(220, 20, as_cmap=True)
    sns.heatmap(corr_mat, square=True, mask=mask, cmap=cmap,
                annot=True, fmt='.2f', vmin=-1, vmax=1,
                annot_kws={'fontsize':'small'},
                ax=ax);
    ax.set_title('Pearson correlation heatmap');
    return corr_mat

corr_mat = plot_corr_matrix(data1)
```

```python
class DataPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self):
        return

    def fit(self, x, y=None): # the fit method acts as a placeholder, it basically does nothing
        return self

    def transform(self, x):
        x = preprocess_raw(x)
        return x
```

```python

class OrdinalEncoder_custom(TransformerMixin, BaseEstimator):
    def __init__(self, categories='auto'):
        self.categories = categories
        self.encoder = OrdinalEncoder(categories=categories)

    def fit(self, x, y=None):
        self.encoder.fit(x, y)
        self.feature_names = x.columns.tolist()
        return self

    def transform(self, x, y=None):
        return self.encoder.transform(x)

    def fit_transform(self, x, y=None):
        self.feature_names = x.columns.tolist()
        return self.encoder.fit_transform(x, y)

    def get_feature_names_out(self):
            return self.feature_names
```

```python
# Numerical preprocessor
num_features = x_train.select_dtypes(include=['int', 'float']).columns.tolist()
num_transformer = Pipeline([('imputer', SimpleImputer(strategy='median')),
                            ('scaler', StandardScaler())])
# Ordinal features
ord_features = ['education_level']
ord_transformer = OrdinalEncoder_custom(categories=[['Unknown', 'Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate']])
# Categorical preprocessor
cat_features = ['gender', 'marital_status', 'income_category', 'card_category']
cat_transformer = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
# Combined preprocessor
preprocessor = ColumnTransformer([('ords', ord_transformer, ord_features),
                                  ('cats', cat_transformer, cat_features),
                                  ('nums', num_transformer, num_features)],
                                 remainder='drop')
```




MODEL BUILDING, TRAINING AND EVALUATION

```python
import time
def evaluate_clf(clf,clf_name):
    start = time.time()
    clf.fit(x_train,y_train);
    end = time.time()

    y_train_pred = clf.predict(x_train)
    y_test_pred = clf.predict(x_test)
    y_test_proba = clf.predict_proba(x_test)
    cv_scores = cross_val_score(clf,x,y, cv=5)

    fig, axs = plt.subplots(1,3)
    fig.set_size_inches(10,4)
    fig.suptitle(f'Name: {clf_name}')
    plt.tight_layout()
    PrecisionRecallDisplay.from_estimator(clf, x_test, y_test, ax=axs[0]);
    RocCurveDisplay.from_estimator(clf, x_test, y_test, ax=axs[1]);
    ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test, ax=axs[2]);

    acc_train = accuracy_score(y_train,y_train_pred)
    acc_test = accuracy_score(y_test,y_test_pred)

    precision = precision_score(y_test, y_test_pred)
    recall = recall_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred)
    auc = roc_auc_score(y_test, y_test_proba[:,1])
    axs[0].set_title('Precision-Recall curve')
    axs[1].set_title('ROC curve')

    eval_dict = {'modelname':clf_name,
                'acc_train':acc_train,
                'acc_test':acc_test,
                'precision':precision,
                'recall':recall,
                'f1':f1,
                'auc':auc,
                'cv_mean':cv_scores.mean(),
                'cv_std':cv_scores.std(),
                'training_time':(end-start)}
    return eval_dict

model_metrics = pd.DataFrame(columns=['modelname', 'acc_train', 'acc_test',
                                      'precision', 'recall', 'f1', 'auc',
                                      'cv_mean', 'cv_std','training_time'])
```

```python
from sklearn.metrics import classification_report

# metrics
model_metrics = pd.DataFrame(columns=['modelname', 'acc_train', 'acc_test',
                                      'precision', 'recall', 'f1', 'auc',
                                      'cv_mean', 'cv_std', 'training_time'])

# log reg
logreg_clf = Pipeline([('preprocessor', preprocessor),
                       ('logreg', LogisticRegression(max_iter=1000))])

# Evaluation
metrics = evaluate_clf(logreg_clf, 'Logistic regression')

y_pred = logreg_clf.fit(x_train, y_train).predict(x_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))

metrics_data = pd.DataFrame([metrics])

model_metrics = pd.concat([model_metrics.round(3), metrics_data], ignore_index=True)
print(model_metrics)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd

categorical_columns = x_train.select_dtypes(include=['object', 'category']).columns
numerical_columns = x_train.select_dtypes(include=['number']).columns

# Define preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_columns),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)
    ],
    remainder='passthrough'  # Keep other columns as-is
)

# Apply preprocessing pipeline
x_train_processed = preprocessor.fit_transform(x_train)
x_test_processed = preprocessor.transform(x_test)

# Converting processed data to dense array (if it is  sparse)
x_train_scaled = x_train_processed.toarray() if hasattr(x_train_processed, "toarray") else x_train_processed
x_test_scaled = x_test_processed.toarray() if hasattr(x_test_processed, "toarray") else x_test_processed

# ann
ann1 = Sequential([
    Dense(64, activation='relu', input_shape=(x_train_scaled.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # Assuming binary classification
])

# compile
ann1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

ann1.fit(x_train_scaled, y_train, epochs=50, batch_size=32)
accuracy = ann1.evaluate(x_test_scaled, y_test)
print(f"Test Accuracy: {accuracy[1] * 100:.2f}%")
```

```python
from sklearn.preprocessing import LabelEncoder
from pytorch_tabnet.tab_model import TabNetClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import time

# Encode categorical columns
label_encoders = {}
for col in x_train.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
    label_encoders[col] = le

# Convert target variable if it's categorical
if y_train.dtype == 'object':
    target_encoder = LabelEncoder()
    y_train = target_encoder.fit_transform(y_train)
    y_test = target_encoder.transform(y_test)

# Define the TabNet model
tabnet_clf = TabNetClassifier()

# Start timer for training
start_time = time.time()

# Train the TabNet model
tabnet_clf.fit(
    X_train=x_train.to_numpy(),
    y_train=y_train,
    eval_set=[(x_test.to_numpy(), y_test)],
    eval_name=["test"],
    eval_metric=["accuracy"],
    max_epochs=50,
    patience=10,
    batch_size=256,
    virtual_batch_size=128,
    num_workers=0,
    drop_last=False
)

# Record training time
training_time = time.time() - start_time

y_train_pred = tabnet_clf.predict(x_train.to_numpy())
y_test_pred = tabnet_clf.predict(x_test.to_numpy())

acc_train = accuracy_score(y_train, y_train_pred)
acc_test = accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred, average='binary')
recall = recall_score(y_test, y_test_pred, average='binary')
f1 = f1_score(y_test, y_test_pred, average='binary')
y_test_pred_proba = tabnet_clf.predict_proba(x_test.to_numpy())[:, 1]
auc = roc_auc_score(y_test, y_test_pred_proba)

metrics = {
    'modelname': 'TabNet',
    'acc_train': acc_train,
    'acc_test': acc_test,
    'precision': precision,
    'recall': recall,
    'f1': f1,
    'auc': auc,
    'training_time': training_time
}

print(metrics)
```
